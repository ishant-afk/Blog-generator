```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
# This will by default set all the tensors to cuda tensor to help better memory allocation
```

```python
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b7")
```

```python
set_seed(424245)
```

```python
blogpost_title = 'The Importance Of Morals in life' # replace this with your own blogpost title
```

```python
prompt = f'This is a blog post on {blogpost_title} \n {blogpost_title}'
prompt
```

```python
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
# Encoding the prompt
```

```python
sample = model.generate(**input_ids, max_length=500, num_beams = 2, num_beam_groups = 2, top_k=1, temperature=0.9, repetition_penalty = 3.0, diversity_penalty = 0.7)
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
